```python
import pandas as pd
import random
```

```python
# Симуляция данных
id_num = [i for i in range(100)]
id_client = [random.randint(10000, 10000000) for i in range(100)]
df = pd.DataFrame(data={'id_num':id_num, 'id_client':id_client})
df.head(2)
```

```python
df[1:5]
```

```python
# Сама функция
def group_customers(n_customers):
  group_dict ={}
  ids = df.id_client.values[:n_customers]

  for id in ids:
    group = sum(list(map(int, str(id))))
    if group in group_dict.keys():
      group_dict[group] = group_dict[group]+1
    else:
      group_dict[group] = 1

  return pd.DataFrame(data={'Group':group_dict.keys(), 'Count':group_dict.values()})
```

```python
group_customers(50).head(5)
```

```python
def group_customers_by_first_id(n_customers, n_first_id):
  group_dict ={}
  ids = df.id_client.values[n_first_id:n_customers+n_first_id]

  for id in ids:
    group = sum(list(map(int, str(id))))
    if group in group_dict.keys():
      group_dict[group] = group_dict[group]+1
    else:
      group_dict[group] = 1

  return pd.DataFrame(data={'Group':group_dict.keys(), 'Count':group_dict.values()})
```

```python
group_customers_by_first_id(3, 20)
```
